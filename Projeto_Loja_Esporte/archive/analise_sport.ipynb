# Analise Venda & Produto

### Top 10 Produtos por Faturamento

```python
df_produtos = df_sport.groupby(['ProductName', 'SubCategory']).agg(
    Faturamento=('SalesAmount', 'sum'),
    Lucro=('Profit', 'sum'),
    Qtd_Vendida=('OrderQuantity', 'sum')
).reset_index()

df_produtos['Margem_%'] = (df_produtos['Lucro'] / df_produtos['Faturamento'] * 100).round(1)
```

```python
top_fat = df_produtos.nlargest(10, 'Faturamento')
top_luc = df_produtos.nlargest(15, 'Lucro')

fig = px.bar(top_fat, x='Faturamento', y='ProductName', 
             orientation='h', color='Margem_%',
             color_continuous_scale='RdYlGn',
             title='Top 10 Produtos por Faturamento')
fig.show()
```

Existe uma relação entre faturamento e margem de lucro, produtos que mais faturaram possuem uma margem maior

### Faturamento e Margem por Categoria

```python
df_margem = df_sport.groupby('Category').agg(
    Faturamento=('SalesAmount', 'sum'),
    Lucro=('Profit', 'sum'),
).reset_index()

df_margem['Margem_%'] = (df_margem['Lucro'] / df_margem['Faturamento'] * 100).round(1)
```

```python
fig = go.Figure()

fig.add_trace(go.Bar(
    x=df_margem['Category'],
    y=df_margem['Faturamento'],
    name='Faturamento',
    marker_color='steelblue'
))

fig.add_trace(go.Scatter(
    x=df_margem['Category'],
    y=df_margem['Margem_%'],
    name='Margem%',
    yaxis='y2',
    mode='lines+markers',
    line=dict(color='orange',width=2)
))

fig.update_layout(
    title='Faturamento e Margem por Categoria',
    yaxis=dict(title='Faturamento (U$D)'),
    yaxis2=dict(title='Margem % ',overlaying='y', side='right'),
    legend=dict(x=0.01, y=0.99)
)

fig.show()
```

Bikes lidera em faturamento absoluto, mas sua margem (40,6%) é idêntica à de Clothing, a categoria com menor receita — indicando que escala não gerou vantagem de rentabilidade. Accessories se destaca como exceção: um faturamento absoluto menor, porém a maior margem entre as três categorias, sugerindo que é o segmento mais eficiente proporcionalmente

### Produtos com Maior Volume X Maior Rentabilidade

```python
df_produtos['Rank_Faturamento'] = df_produtos['Faturamento'].rank(ascending=False).astype(int)
df_produtos['Rank_Lucro'] = df_produtos['Lucro'].rank(ascending=False).astype(int)
df_produtos['Diferenca_Rank'] = df_produtos['Rank_Faturamento'] - df_produtos['Rank_Lucro']
```

```python
fig = px.scatter(
    df_produtos,
    x='Faturamento',
    y='Margem_%',
    size='Qtd_Vendida',
    color='SubCategory',
    hover_name='ProductName',
    title='Volume x Rentabilidade por Produto'
)
fig.add_hline(y=df_produtos['Margem_%'].mean(), line_dash='dash', line_color='gray',
              annotation_text='Margem Média')
fig.add_vline(x=df_produtos['Faturamento'].mean(), line_dash='dash', line_color='gray',
              annotation_text='Faturamento Médio')
fig.show()
```

Os produtos mais rentáveis não são os mais vendidos — alto volume de vendas não garante alta margem, o que indica que a estratégia de mix de produtos deve priorizar rentabilidade, não apenas giro.

### Faturamento e Lucro por País

```python
df_pais = df_sport.groupby('Country').agg(
    Faturamento=('SalesAmount', 'sum'),
    Lucro=('Profit', 'sum'),
).reset_index()
df_pais['Margem_%'] = (df_pais['Lucro'] / df_pais['Faturamento'] * 100).round(1)
df_pais = df_pais.sort_values('Faturamento', ascending = False)
```

```python
fig = px.bar(
    df_pais,
    x = 'Country',
    y = ['Faturamento', 'Lucro'],
    barmode = 'group',
    title = 'Faturamento e Lucro por País',
    text_auto= '.2s'
)
fig.show()
```

Estados Unidos e Austrália lideram em faturamento, porém apresentam as menores margens de lucro entre os países analisados. Alemanha e França, apesar do menor volume de receita, demonstram melhor eficiência — maior retorno proporcional sobre o faturamento.

### Participação de Faturamento por País

```python
fig = px.pie(
    df_pais,
    names = 'Country',
    values = 'Faturamento',
    title = 'Participação de Faturamento por País',
    hole = 0.6
)

fig.show()
```

Estados Unidos e Austrália concentram 62,9% do faturamento total, enquanto os quatro países restantes dividem os 37,1% restantes — evidenciando uma distribuição geográfica desigual da receita

### Evolução Anual do Faturamento por País

### Adicionando coluna Ano

```python
df_sport['Year'] = df_sport['OrderDate'].dt.year
```

```python
df_anual = df_sport.groupby(['Year','Country'])['SalesAmount'].sum().reset_index()
```

```python
fig = px.line(
    df_anual,
    x = 'Year',
    y = 'SalesAmount',
    color = 'Country',
    markers = True,
    title = 'Evolução Anual do Faturamento por País',
    labels = {'SalesAmount': 'Faturamento (USD)', 'Year': 'Ano'}
)

fig.update_layout(
    xaxis = dict(tickmode = 'array', tickvals = [2014, 2015, 2016])
)

fig.show()
```

No periodo apresentado nota-se que houve uma queda no faturamento geral de todos os paises no ano de 2015, e que no periodo entre 2015 - 2016 a Australia que liderava com o maior faturamento foi ultrapassada pelo United States, que se tornou o país com maior faturamento em 2016, já o Canada que em 2014 era o quarto maior faturamento, apresentou um movimento negativo, sinalizando uma deterioração consistente no mercado canadense  

# ANALISE TEMPORAL

### Evolução Mensal do Faturamento

```python
df_mensal = df_sport.groupby('Month')['SalesAmount'].sum().reset_index()

fig = px.line(
    df_mensal,
    x = 'Month',
    y = 'SalesAmount',
    markers = True,
    title = 'Evolução Mensal do Faturamento',
    labels = {'SalesAmount': 'Faturamento (USD)', 'Month': 'Período'}
)

fig.update_layout(
    xaxis=dict(
        tickmode = 'array',
        tickvals = df_mensal['Month'][::3].tolist()
    )
)
fig.show()
```

O faturamento apresenta tendência crescente ao longo do ano, com um pico intermediário em junho seguido de correção em julho — que ainda supera os meses anteriores. A trajetória se mantém ascendente até dezembro, onde atinge o maior valor do período, evidenciando forte sazonalidade no final do ano

### Mês que Mais Faturou 

```python
df_mes = df_sport.groupby(['Year','Month'])['SalesAmount'].sum().reset_index()
df_mes['Month_Nome'] = df_mes['Month'].apply(lambda x: pd.Timestamp(2024, x, 1).strftime('%B'))

melhor_mes = df_mes.loc[df_mes['SalesAmount'].idxmax()]
print(f"Melhor mês: {melhor_mes['Month_Nome']}/{melhor_mes['Year']} - Faturamento: ${melhor_mes['SalesAmount']:,.2f}")
```

### Meses que Mais Faturaram por Ano

```python
df_melhor_mes = df_sport.groupby(['Year','Month'])['SalesAmount'].sum().reset_index()
df_melhor_mes['Month_Nome'] = df_melhor_mes['Month'].apply(lambda x: pd.Timestamp(2024, x, 1).strftime('%B'))

resultado = df_melhor_mes.loc[df_melhor_mes.groupby('Year')['SalesAmount'].idxmax()]
print(resultado[['Year','Month_Nome','SalesAmount']].to_string(index = False))
```

```python
fig = px.bar(
    df_melhor_mes,
    x = 'Month_Nome',
    y = 'SalesAmount',
    color = 'Year',
    barmode = 'group',
    title = 'Faturamento Mensal por Ano',
    labels = {'SalesAmount': 'Faturamento (USD)','Month_Nome':'Mês','Year': 'Ano'}
)

fig.show()
```

2016 foi o ano com maior faturamento, superando com folga os anos anteriores em todos os meses. Em contraste, 2015 apresentou queda em relação a 2014 em praticamente todos os meses, configurando um ciclo de retração seguido de forte recuperação em 2016

### Top 10 Dias com Maior Volume de Vendas

```python
df_dia = (df_sport.groupby('OrderDate')['SalesAmount'].sum().reset_index().sort_values('SalesAmount',ascending=False).head(10))

df_dia['OrderDate'] = df_dia['OrderDate'].dt.strftime('%d-%m-%Y')
df_dia['SalesAmount_fmt'] = df_dia['SalesAmount'].apply(lambda x: f'${x:^,.0f}')
df_dia = df_dia.sort_values('SalesAmount', ascending = True)
```

```python
fig = px.bar(
    df_dia,
    x = 'SalesAmount',
    y = 'OrderDate',
    orientation = 'h',
    text = 'SalesAmount_fmt',
    title = 'Top 10 Dias com Maior Volume de Vendas',
    labels = {'SalesAmount': 'Faturamento (USD)', 'OrderDate': 'Data'},
    color='SalesAmount',
    color_continuous_scale = 'Blues'
)

fig.update_traces(textposition = 'outside')

fig.update_layout(
    coloraxis_showscale = False,
    xaxis = dict(tickformat = ',.0f')
)

fig.show()
```

O ranking revela dois padrões de concentração de vendas: os maiores volumes ocorrem nos últimos dias do mês — influenciados pelo ciclo de recebimento de salários — e oito das dez datas são do último trimestre do ano, refletindo o impacto da sazonalidade de fim de ano no comportamento de compra
